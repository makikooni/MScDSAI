## Task 4:
Run a statistical analysis for testing the following hypothesis:

**Hypothesis 2:** A woman’s tendency to participate actively in the conversation correlates positively with the number of females in the discussion.

```python
# Your code here
df1 = df[["ThreadId","Female_Contributions","UniqueFemaleContributors","Female"]]
df1.head()
```

```python
df2 = df1[df1["Female_Contributions"]==df1["UniqueFemaleContributors"]].copy()
```

```python
len(df2)
```

```python
df2["cum_fem_participants"] = df2.groupby("ThreadId")["Female"].cumsum()  
```

```python
df2["next_is_female"] = df2.groupby("ThreadId")["Female"].shift(-1)  
```
